### calculate  Monthly recovery  ratios for the US

here we have to calculate date by month so we extract month by year from the date column

using new fucntion .dt.to_period('M'), to get year-month format

```python
#df_final[df_final['Country/Region'] == "US"]
df_us = df_final[df_final['Country/Region'] == "US"]
df_us['yyyy-mm'] = df_us['Date'].dt.to_period('M')
df_us = df_us.groupby('yyyy-mm')[['recover_cases','confirmed_cases']].sum()
df_us['recover_rate'] = df_us['recover_cases']/df_us['confirmed_cases']
df_us['recover_rate'].plot()
```
