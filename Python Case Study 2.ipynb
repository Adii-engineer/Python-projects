# Task 3: Data Quality Checks

Compairing 2 dataset values on state level to see accuracy

```python
state_agg = state_and_users.groupby(['State'])[['Transactions','Amount (INR)','Registered Users']].sum().reset_index()
state_agg.columns = ['State_State','State_Transactions','State_Amount','State_Registered_Users']
state_agg
district_agg = district_and_users.groupby(['State'])[['Transactions','Amount (INR)','Registered Users']].sum().reset_index()
district_agg.columns = ['distric_State','distric_Transactions','distric_Amount','distric_Registered_Users']
district_agg
```

##### compairing the values

```python
merged = state_agg.merge ( district_agg , left_on = 'State_State' , right_on = 'distric_State' , how = 'outer' )
```

```python
print('Transactions Match =',(merged['distric_Transactions'] == merged['State_Transactions']).sum()/merged.shape[0]*100,'%')
print('Amount Match =',(merged['distric_Amount'] == merged['State_Amount']).sum()/merged.shape[0]*100,'%')
print('Registered Users Match =',(merged['distric_Registered_Users'] == merged['State_Registered_Users']).sum()/merged.shape[0]*100,'%')
```

# Task 5: Data Visualization

Plotting total transactions and amount over time for perticular state

```python
trend = state_and_users.groupby(['State','Year','Quarter'])[['Transactions','Amount (INR)']].sum().reset_index()
trend = trend[trend['State']=='Delhi']

# sns.lineplot(x = 'Year',y='Transactions',data=trend, color='tab:blue', marker='o', label='Transactions')
# sns.lineplot(x = 'Year',y='Amount (INR)',data=trend, color='tab:blue', marker='s', label='Amount',hue='Quarter').set_title("Trend for Delhi")

# Plot the data
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot Transactions on the primary y-axis
ax1.set_xlabel('Year')
ax1.set_ylabel('Transactions', color='tab:blue')
ax1.plot(trend['Year'], trend['Transactions'], color='tab:blue', marker='o', label='Transactions')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Create a secondary y-axis to plot Amount
ax2 = ax1.twinx()
ax2.set_ylabel('Amount', color='tab:red')
ax2.plot(trend['Year'], trend['Amount (INR)'], color='tab:red', marker='o', label='Amount')
ax2.tick_params(axis='y', labelcolor='tab:red')

# Add a title and grid
# plt.title('Transactions and Amount Over the Years for Delhi')
fig.tight_layout()
plt.grid(True)
plt.show()
```
